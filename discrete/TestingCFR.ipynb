```python
from MCTS import MCTS
import numpy as np
```

```python
np.random.seed(0)
X = np.random.rand(1000, 100)
coefficients = np.random.rand(100)
noise = np.random.normal(scale=5.0, size=(1000,))
y = X.dot(coefficients) + noise
```

```python
tree = MCTS((X, y))
```

```python
tree.build_tree(iterations=int(1e7))
```
